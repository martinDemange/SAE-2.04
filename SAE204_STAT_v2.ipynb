```python
import pandas as pa
import numpy as np
import matplotlib.pyplot as plt

# Read the csv file so they can be used
bien = pa.read_csv("./BIEN.csv")
proprio = pa.read_csv("./PROPRIO.csv")
espace = pa.read_csv("./ESPACE.csv")
```

### Prix moyen du m² par ville
Nous avons decidé de montrer le prix du m² moyen par ville afin de determine dans quelle ville les biens que nous possédons sont les plus cher.

```python
from IPython.display import display

# Fait une jointure sur BIEN et ESPACE, avec comme condition que IDEBIEN doit etre egale à IDE
# Requete SQL correpondante à la jointure :  
# SELECT *
# FROM ESPACE E, BIEN B
# WHERE B.IDEBIEN = E.IDE;
bien_espace = pa.merge(bien, espace, left_on='IDEBIEN', right_on='IDE')

# fait un groupby afin de range les bien en fonction de la ville ou il sont situe
stat = bien_espace.groupby("VILLE")

# Calcule le prit du m² pour tout les bien en divisant le prix par la surface
stat1 = stat.apply( lambda x: x["PRIX"] / x["SUPERFICIE"] )
# calcule la moyenne des prix du m² par ville
price_surface_city = stat1.groupby("VILLE").mean() 

# affiche le resultat dans un diagramme en bande
plt.figure()
price_surface_city.plot(kind='bar', color='blue', edgecolor='black')
plt.title('Prix moyen du m² de nos bien immobilier par ville')
plt.ylabel('Prix moyen du m²')
plt.xticks(rotation=0)
plt.grid(visible=True, which='major', axis='y')
plt.show()


```

Le diagramme en bande permet une comparaison facilement du prix moyen du m² ( axe y ) en fonction de la ville etudier ( axe x ).
